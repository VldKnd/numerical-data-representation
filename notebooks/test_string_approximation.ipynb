```python
from source.numeric_representation import MinilmEmbedding, LoagrithmicMinilmEmbedding, SinusoidalMinilmEmbedding, SigmoidMinilmEmbedding
from source.utils import pairwise_cosine_similarity_matrix
import torch
import matplotlib.pyplot as plt

sentences: list[str | int | float] = [
    "1",
    42,
    33.5,
    "45",
    "Sentence without embedding",
    "My favorite number is 42",
    "The answer to all is 42",
    "There are 42 rabbits in the bag",
    "42 parrots are flying in the sky",
    "42 rabbits are running in the yard",
]
```

```python
model = LoagrithmicMinilmEmbedding()
embeddings = model.encode(input=sentences)

pairwise_l2_distances = torch.cdist(embeddings.unsqueeze(0), embeddings.unsqueeze(0)).squeeze(0)
pairwise_l2_similarities = torch.exp(-pairwise_l2_distances)
pairwise_cosine_similarities = pairwise_cosine_similarity_matrix(embeddings, embeddings)

figure = plt.figure(figsize=(10, 15))

plt.subplot(2, 1, 1)
im = plt.imshow(pairwise_l2_similarities)
plt.xticks(
    torch.arange(len(sentences)),
    labels=[str(sentence) for sentence in sentences],
    rotation=45,
    ha="right",
    rotation_mode="anchor"
)
plt.yticks(torch.arange(len(sentences)), labels=[str(sentence) for sentence in sentences])

for i in range(len(sentences)):
    for j in range(len(sentences)):
        text = plt.text(j, i, f"{pairwise_l2_similarities[i, j].item():.9f}"[:4],
                       ha="center", va="center", color="w")

plt.title("Pairwise distance between sentences")
plt.tight_layout()

plt.subplot(2, 1, 2)
im = plt.imshow(pairwise_cosine_similarities)
plt.xticks(
    torch.arange(len(sentences)),
    labels=[str(sentence) for sentence in sentences],
    rotation=45,
    ha="right",
    rotation_mode="anchor"
)
plt.yticks(torch.arange(len(sentences)), labels=[str(sentence) for sentence in sentences])

for i in range(len(sentences)):
    for j in range(len(sentences)):
        text = plt.text(j, i, f"{pairwise_cosine_similarities[i, j].item():.9f}"[:4],
                       ha="center", va="center", color="w")

plt.title("Cosine similarity between embeddings")
plt.tight_layout()
plt.show()
```
